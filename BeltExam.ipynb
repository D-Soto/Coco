```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
EmpData_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Project/employee_attrition.csv')
EmpData_df.head()
```

```python
EmpData_df.info()
```

```python
EmpData_df.groupby('Gender').count()
```

```python
FullData_df = EmpData_df.fillna('Male')
FullData_df
#only filled Male because only Females were counted
```

```python
FullData_df.groupby('YearsAtCompany').count()
```

```python
FullData_df['YearsAtCompany'] = FullData_df['YearsAtCompany'].str.replace('!','').astype(int)
# this code worked to change the object row to an int
```

```python
FullData_df.info()
```

```python
FullData_df.set_index('EmployeeNumber')
```

```python
Sat_Fillter = FullData_df.groupby(['JobRole'])[['JobSatisfaction']].mean()
Sat_Fillter
```

```python
Sat_Fillter.plot(kind = 'bar' , figsize = (20,10))
plt.xlabel('JobRole')
plt.ylabel('JobSatisfaction')
plt.title('Average Satisfaction in Job Roles')
plt.legend(loc='upper right',)
plt.show()
```

```python
#Job satification is sitting a little over half
```

```python
Pro_Fillter = FullData_df.groupby(['JobRole'])[['MonthlyIncome']].mean()
Pro_Fillter
```

```python
Pro_Fillter.plot(kind = 'bar' , figsize = (20,10))
plt.xlabel('JobRole')
plt.ylabel('MonthlyIncome')
plt.title('Pay Splits in Job Roles')
plt.legend(loc='upper right',)
plt.show()
```

```python
#the avrage pay splits between jobs
```
